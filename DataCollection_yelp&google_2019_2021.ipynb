```python
### Code referenced from a very interesting blog by Lee Robinson, https://leerob.io/blog/google-yelp-foursquare-apis-python

lifeline_list = []

       
for category in categories2021:
        
    for i in range(4):
        
        # Yelp Request request
        url = 'https://api.yelp.com/v3/businesses/search'
        key = '' # the key is removed for security purpose
        headers = {'Authorization': 'bearer %s' % key}
            
        parameters = {'categories': category,
                        'location': 'Houston',
                        'radius': 40000,
                        'limit': 50,
                        'offset': 50}
                        #  'open_at': 1588962191
                        # 'term': 'Houston Police Department Headquarters'}

        resp = requests.get(url=url,headers=headers, params=parameters)
        data = json.loads(resp.text)
        #data['businesses']
    
#            # data exploration
        for i in range(len(data['businesses'])):
            store_dict2 = {}
            store_dict2['Business_name'] = data['businesses'][i]['name']
            store_dict2['latitude'] = data['businesses'][i]['coordinates']['latitude']
            store_dict2['longitude'] = data['businesses'][i]['coordinates']['longitude']
            store_dict2['category'] = category
           # store_dict2['zipcode'] = data['businesses'][i]['location']['zip_code']
            lifeline_list.append(store_dict2)
            
    
lifeline_df1 = pd.DataFrame(lifeline_list)
lifeline_df1['lifeline_num'] = lifeline_df1['category'].map(lifelines_1to7)
lifeline_df1['lifeline_name'] = lifeline_df1['lifeline_num'].map(lifeline_names)
lifeline_df1['Source'] = 'Yelp'
```

```python
### Code inspred from GA_DSI8_FEMA_Lifelines

def scrape_google(categorylist, api_key):
    
    names = []
    categories = []
    lat = []
    long = []
    lifeline = []
    lifelinename = []
    
    ## Extract from Google Places
    for category in categorylist:
        query_result = api_key.nearby_search(location='Houston, Texas', keyword=category,radius=40000)

        for place in query_result.places:
            names.append(place.name)
            categories.append(category)
            lat.append(place.geo_location['lat'])
            long.append(place.geo_location['lng'])
 
        print(f'Category Check: {category}')
        time.sleep(2)

    df = pd.DataFrame(columns=['Business_name'])
    df['Business_name'] = names
    df['latitude'] = lat
    df['longitude'] = long
    df['category'] = categories
    df['lifeline_num'] = df['category'].map(categories2021_google)
    df['lifeline_name'] =  df['lifeline_num'].map(lifeline_names)
    df['Source'] = 'Google'

   # df['Lifeline'] = lifeline
    
    return df
```
